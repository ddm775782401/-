```python
import numpy as np
from matplotlib import pyplot as pl
from mpl_toolkits.mplot3d import Axes3D as ax3
from sklearn.metrics import mean_squared_error
class RBF_1D:
    def createdata(self,X,Y,basicname='MQ',method='pinv',*args):
        #X,Y为拟合数据，basicname为基函数选择：MQ，Gaussian，Poly
        #method为解方程的方法：inv，pinv
        #args为基函数的参数
        X=numpy.array(X)
        Y=numpy.array(Y)
        n=len(X)
        X=X.reshape((n,1))
        Y=Y.reshape((n,1))
        #basic1为MQ基函数，参数c，beta
        #basic2为高斯基函数，参数a
        #basic3为多项式基函数，参数N
        c,beta=args
        a=args[0]
        N=args[0]
        basic1=lambda x:(x-c**2)**beta
        basic2=lambda x:numpy.exp(-numpy.power(x/a,2))
        basic3=lambda x:x**N
        basicfunction={'MQ':basic1,'Gaussian':basic2,'Poly':basic3}
        basic=basicfunction[basicname]

        A=numpy.zeros((n,n))
        XX=numpy.zeros((n,n))
        F=numpy.zeros((n,1))
        a=numpy.zeros((n,1))
        for i in range(n):
            for j in range(n):
                XX[i,j]=(X[i,0]-X[j,0])**2
                A[i,j]=basic(XX[i,j])
        F=Y.copy()
        if method=='inv':
            a=numpy.dot(numpy.linalg.inv(A),F)
        else:
            a=numpy.dot(numpy.linalg.pinv(A),F)
        self.x=X
        self.y=Y
        self.a=a
        self.basic=basic
    def predict(self,x0):
        x0=numpy.array(x0)
        x0.reshape((len(x0),1))
        n1=len(self.x)
        n2=len(x0)
        A=numpy.zeros((n2,n1))
        for i in range(n2):
            for j in range(n1):
                ans=(x0[i]-self.x[j])
                ans=ans**2
                A[i,j]=self.basic(ans)
        F=numpy.dot(A,self.a)
        return F
    def show(self):
        #展示逼近情况并返回误差向量
        figure1=pl.figure()
        pl.plot(self.x,self.y,'y<',label='origin data')
        y=self.predict(self.x)
        error=mean_squared_error(y, self.y)
        pl.plot(self.x,y,label='fitting data')
        pl.legend()
        figure1.show()
        return error
```

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

x=numpy.arange(-6*math.pi,6*math.pi,0.1)
y=numpy.sin(x)
R=RBF_1D()
R.createdata(x,y,'Poly','pinv',0.5,3)
error=R.show()
x0=[0]
y0=R.predict(x0)
print(y0)
print ('均方误差:',error)
plt.text(18,0.5,'Loss='+str(error),fontdict={'size':20,'color':'red'})
```

# sinx/x

```python
x=np.arange(0.0001,6*math.pi,0.1)
y=np.sin(x)/np.abs(x)
R=RBF_1D()
R.createdata(x,y,'Poly','pinv',0.5,3)
error=R.show()
x0=[0]
y0=R.predict(x0)
print(y0)
print ('均方误差:',error)
plt.text(18,0.5,'Loss='+str(error),fontdict={'size':20,'color':'red'})
```

# 1/X^3

```python
rng = np.random.RandomState(0)
x=np.logspace(-1,1,300)
y=1/x**3
y[::50] += 2 * (0.5 - rng.rand(int(x.shape[0]/50)))
R=RBF_1D()
R.createdata(x,y,'Poly','pinv',0.5,3)
error=R.show()
x0=[0]
y0=R.predict(x0)
print(y0)
print ('均方误差:',error)
plt.text(4,200,'Loss='+str(error),fontdict={'size':20,'color':'red'})
```

```python
x_t=np.logspace(-1,1,50)
x_t=x_t.reshape((-1,1))
y_t=(1/x_t**3).ravel()
y_tp=R.predict(x_t)
test_error=mean_squared_error(y_t, y_tp)
test_error
```
